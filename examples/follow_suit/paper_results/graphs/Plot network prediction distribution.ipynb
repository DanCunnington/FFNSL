```python
import numpy as np
import torch
import pandas as pd
import json
import math
import matplotlib.pyplot as plt
```

```python
networks = ['softmax', 'edl_gen']
noise_pct = 0
repeats = list(range(1,51))
decks = ['standard', 'batman_joker', 'captain_america', 'adversarial_standard',
         'adversarial_batman_joker', 'adversarial_captain_america']
```

```python
def plot_rank_results(d, n, noise_pct):
    fig = plt.figure(constrained_layout=True, figsize=(4,4))
    axes = plt.gca()
 
    with open('../nsl/network_predictions/{0}/{1}_more_repeats.json'.format(n,d),'r') as rf:
        rf = json.loads(rf.read())
        np_res = rf['noise_pct_{0}'.format(noise_pct)]
        rd = np_res['rank_dist']
        if n =='softmax':
            color = 'tab:blue'
            label = 'Softmax'
        else:
            color = 'tab:orange'
            label = 'EDL-GEN'

        plt.bar(list(rd.keys()), rd.values(), color=color,label=label)
    axes.set_xlabel('Playing card rank')
    axes.set_ylabel('Distribution')
    axes.set_ylim([0,0.8])
#     axes.legend()
    plt.savefig('network_prediction_distribution/{0}_{1}_rank_noise_pct_{2}_95_100.pdf'.format(d,n,noise_pct), format='pdf', bbox_inches='tight')
```

```python
def plot_suit_results(d, noise_pct):
    fig = plt.figure(constrained_layout=True, figsize=(4,4))
    axes = plt.gca()
    for n in ['edl_gen']:
        with open('../nsl/network_predictions/{0}/{1}_more_repeats.json'.format(n,d),'r') as rf:
            rf = json.loads(rf.read())
            np_res = rf['noise_pct_{0}'.format(noise_pct)]
            sd = np_res['suit_dist']
            if n =='softmax':
                color = 'b'
                label = 'Softmax'
            else:
                color = 'k'
                label = 'EDL-GEN'

            plt.bar(list(sd.keys()), sd.values(), color=color,label=label,alpha=0.5)
    axes.set_xlabel('Playing card suit')
    axes.set_ylabel('Distribution')
    axes.set_ylim([0,0.8])
    axes.legend()
```

```python
plot_rank_results('adversarial_captain_america','softmax',95)
```
