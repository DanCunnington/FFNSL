```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(exp):
    large = json.load(open('../cnn/large/standard.json', 'r'))
    results = large['noise_pct_0']['training_losses'][str(exp)] 
    x = []
    for idx in list(range(len(results))):
        x.append(idx*6400)
    # Generate Training Graph
    fig, ax1 = plt.subplots()

    color = 'red'
    ax1.set_xlabel('Number of training examples observed')
    ax1.set_ylabel('Binary Cross-Entropy Loss', color=color)
    ax1.plot(x, results, color=color, label='Train Loss', zorder=1)
    ax1.tick_params(axis='y', labelcolor=color)
    lgd = fig.legend(loc="upper right", bbox_to_anchor=(1.25, 1))
```

```python
def plot_nine_by_nine(res):
    x = []
    for idx in list(range(len(res['1']))):
        x.append(idx*6400)
    # Generate Training Graph
    fig, ax1 = plt.subplots()

    color = 'red'
    ax1.set_xlabel('Number of training examples observed')
    ax1.set_ylabel('Binary Cross-Entropy Loss', color=color)
    ax1.plot(x, res['1'], color=color, label='Train Loss', zorder=1)
    ax1.tick_params(axis='y', labelcolor=color)
    lgd = fig.legend(loc="upper right", bbox_to_anchor=(1.25, 1))
```
